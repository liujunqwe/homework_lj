## 问题 1  

> 给定一个已排序的链表的头 `head` ， *删除所有重复的元素，使每个元素只出现一次* 。返回 *已排序的链表* 。链表的类如下所示：

```python
class NodeList:
    def __init__(self, val=None, right=None):
        self.val   = val
        self.right = right
```

输入是一个数组，你首先需要将数组转化为链表，然后删除链表中的重复元素，再遍历链表元素，以一个数组的形式返回。请设计一个算法解决上述任务，分析算法设计思路，计算时间复杂度, 并基于python编程实现。

e.g.  输入：head=[1, 1, 2, 3, 3]   输出：[1, 2, 3]

### 算法设计思路

解决这个问题可以分为三个主要步骤：

1.  **将输入数组转换为排序链表**：
    *   首先，检查输入数组是否为空。如果为空，则返回一个空列表或表示空链表的适当值（根据题目是返回数组，所以最终是空数组）。
    *   如果数组不为空，创建链表的头节点，其值为数组的第一个元素。
    *   使用一个指针（例如 `tail`）跟踪链表的最后一个节点，初始时指向头节点。
    *   遍历输入数组从第二个元素开始的剩余部分。对于每个元素，创建一个新的 `NodeList` 节点，并将其设置为当前 `tail` 节点的 `right` 指针。然后，更新 `tail` 指针到这个新添加的节点。

2.  **删除已排序链表中的重复元素**：
    *   由于链表是已排序的，所有重复的元素都将是相邻的。
    *   使用一个指针 `current` 初始化为链表的头节点 `head`。
    *   当 `current` 和 `current.right` (即下一个节点) 都存在时，进行迭代：
        *   比较 `current.val` 和 `current.right.val`。
        *   如果它们的值相等，说明 `current.right` 是一个重复节点。为了删除它，我们将 `current.right` 指针指向 `current.right.right`，从而跳过（并实际上是删除了从链表中）重复的节点。注意，此时 `current` 指针不移动，因为新的 `current.right` 可能是另一个重复项。
        *   如果它们的值不相等，说明 `current` 的下一个节点不是重复的，所以我们将 `current` 指针移动到 `current.right`。
    *   遍历结束后，链表中将没有重复元素。

3.  **将处理后的链表转换回数组**：
    *   创建一个空数组用于存放结果。
    *   使用一个指针（例如 `temp`）从处理后的链表的头节点开始遍历。
    *   对于链表中的每个节点，将其 `val` 添加到结果数组中。
    *   移动 `temp` 指针到其 `right` 节点，直到遍历完整个链表。
    *   返回结果数组。

### 时间复杂度分析

*   **数组转换为链表**：此步骤需要遍历输入数组一次，为数组中的每个元素创建一个节点。如果数组的长度为 N，则此操作的时间复杂度为 O(N)。
*   **删除重复元素**：此步骤需要遍历链表一次。在最坏的情况下（没有重复元素或所有元素都只重复一次），每个节点都会被访问一次。如果存在连续的重复元素，`current` 指针在删除重复元素时保持不变，但 `current.right` 指针会向前移动。总体来说，每个节点关系（`right` 指针）只会被调整或遍历一次。因此，时间复杂度为 O(N)。
*   **链表转换回数组**：此步骤需要遍历去重后的链表一次。如果去重后的链表长度为 M (M ≤ N)，则时间复杂度为 O(M)。

综合来看，整个算法的总时间复杂度是 O(N) + O(N) + O(M)，由于 M ≤ N，所以最终时间复杂度为 **O(N)**。

idea：

````python
# add your idea here
class NodeList:
    def __init__(self, val=None, right=None):
        self.val   = val
        self.right = right

def array_to_linked_list(arr):
    """将数组转换为链表"""
    if not arr:
        return None
    head = NodeList(arr[0])
    current = head
    for i in range(1, len(arr)):
        current.right = NodeList(arr[i])
        current = current.right
    return head

def linked_list_to_array(head):
    """将链表转换为数组"""
    arr = []
    current = head
    while current:
        arr.append(current.val)
        current = current.right
    return arr

def delete_duplicates_in_sorted_list(head):
    """删除已排序链表中的重复元素"""
    if not head:
        return None
    
    current = head
    while current and current.right:
        if current.val == current.right.val:
            current.right = current.right.right  # 跳过重复的节点
        else:
            current = current.right  # 移动到下一个不同的节点
    return head

def process_sorted_list_array(input_arr):
    """
    主函数：将数组转换为链表，删除重复项，然后转换回数组。
    """
    # 1. 将数组转换为链表
    linked_list_head = array_to_linked_list(input_arr)
    
    # 2. 删除链表中的重复元素
    processed_list_head = delete_duplicates_in_sorted_list(linked_list_head)
    
    # 3. 将处理后的链表转换回数组
    result_array = linked_list_to_array(processed_list_head)
    
    return result_array

# 示例
head_arr1 = [1, 1, 2, 3, 3]
output1 = process_sorted_list_array(head_arr1)
print(f"输入: {head_arr1}, 输出: {output1}") # 期望输出: [1, 2, 3]

head_arr2 = [1, 1, 1, 2, 2]
output2 = process_sorted_list_array(head_arr2)
print(f"输入: {head_arr2}, 输出: {output2}") # 期望输出: [1, 2]

head_arr3 = []
output3 = process_sorted_list_array(head_arr3)
print(f"输入: {head_arr3}, 输出: {output3}") # 期望输出: []

head_arr4 = [1, 2, 3, 4, 5]
output4 = process_sorted_list_array(head_arr4)
print(f"输入: {head_arr4}, 输出: {output4}") # 期望输出: [1, 2, 3, 4, 5]

# 算法时间复杂度已在上面的Markdown中分析为: O(N)
````

> 下面是一个经典的算法问题：

>
> 由于要多次查找数组中元素的位置，为了提高查询效率可以使用哈希表来存储数组中的数据，在哈希表中查询一个元素的复杂度为O(1)。 已知python中的字典是使用哈希表实现的，即使用`dict[key]`查询对应的value时间复杂度为O(1), python提供了查询字典是否包含某个key的功能：`key in dict`，时间复杂度也是O(1)

请根据上面信息，设计一个时间复杂度为O(n) 的算法，解决上述算法问题

e.g.   

输入：nums=[2,7,11,15], target=9， 输出：[0，1]

输入：nums=[3,2,4], target=6, 输出：[1,2]

输入：nums=[3,3], target=6,  输出：[0,1]


````python
# add your idea here
def two_sum(nums, target):
    """
    给定一个整数数组 nums 和一个整数目标值 target，
    请你在该数组中找出 和为目标值 target 的那 两个 整数，并返回它们的数组下标。

    算法思路：
    1. 初始化一个空的哈希表 (字典) `num_map`，用于存储数组中每个数字及其对应的下标。
       格式：{数字: 下标}
    2. 遍历数组 `nums` 中的每个元素（同时获取其值 `num` 和下标 `i`）。
    3. 对于每个元素 `num`，计算需要找到的另一个数 `complement = target - num`。
    4. 检查 `complement` 是否已经存在于 `num_map` 中：
        a. 如果 `complement` 存在于 `num_map` 中，说明我们已经找到了和为 `target` 的两个数。
           这两个数分别是当前的 `num` (其下标为 `i`) 和 `complement` (其下标存储在 `num_map[complement]`)。
           题目保证每种输入只会对应一个答案，所以可以直接返回 `[num_map[complement], i]`。
        b. 如果 `complement` 不存在于 `num_map` 中，说明到目前为止还没有找到与当前 `num` 配对的数。
           此时，将当前的 `num` 及其下标 `i` 存入 `num_map` 中，即 `num_map[num] = i`。
           这样做是为了后续的元素在查找其 `complement` 时，能够找到当前这个 `num`。
    5. 如果遍历完整个数组都没有找到符合条件的两个数（根据题目假设，这种情况不会发生，因为总有答案），
       可以返回一个空列表或抛出异常，但根据题意，总能找到。

    时间复杂度分析：
    - 我们只需要遍历数组一次。对于数组中的每个元素，我们执行以下操作：
        - 计算补数：O(1)
        - 在哈希表中查找补数是否存在：平均情况下为 O(1)。
        - 将当前元素及其下标存入哈希表：平均情况下为 O(1)。
    - 由于对N个元素中的每一个都执行常数时间的操作，所以总的时间复杂度是 O(N)。

    空间复杂度分析：
    - 在最坏的情况下（例如，没有两个数字的和等于目标值，或者目标值需要数组末尾的数字才能构成），
      哈希表可能需要存储接近N个元素。
    - 因此，空间复杂度是 O(N)。
    """
    num_map = {}  # 哈希表，用于存储数字及其下标
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            # 如果补数在哈希表中，说明找到了两个数
            return [num_map[complement], i]
        # 如果补数不在哈希表中，将当前数和它的下标存入
        num_map[num] = i
    
    # 根据题目假设，总是存在答案，所以理论上不会执行到这里
    return [] 

# 示例
nums1 = [2, 7, 11, 15]
target1 = 9
output1 = two_sum(nums1, target1)
print(f"输入: nums={nums1}, target={target1}, 输出: {output1}") # 期望输出: [0, 1]

nums2 = [3, 2, 4]
target2 = 6
output2 = two_sum(nums2, target2)
print(f"输入: nums={nums2}, target={target2}, 输出: {output2}") # 期望输出: [1, 2]

nums3 = [3, 3]
target3 = 6
output3 = two_sum(nums3, target3)
print(f"输入: nums={nums3}, target={target3}, 输出: {output3}") # 期望输出: [0, 1]

# your algorithm time complexity is: O(n)
# your algorithm time complexity is: O(n)
````

## 问题 3:   

> 栈是一种常用的数据结构，编译器中通常使用栈来实现表达式求值。
>
> 以表达式 $3+5 \times 8-6$​ 为例。编译器使用两个栈来完成运算，即一个栈保持操作数，另一个栈保存运算符。
>
> 1. 从左向右遍历表达式，遇到数字就压入操作数栈；
>
> 2. 遇到运算符，就与运算符栈的栈顶元素进行比较。如果比运算符栈顶元素的优先级高，就将当前运算符压入栈；如果比运算符栈顶元素的优先级低或者相同，从运算符栈中取栈顶运算符，从操作数栈的栈顶取 2 个操作数，然后进行计算，再把计算完的结果压入操作数栈，继续比较。

根据上述原理，请设计一个算法完成表达式的运算，当输入为表达式字符串，返回对应的计算结果。分析算法设计思路，计算时间复杂度，并基于python编程实现

**note：**

1. 假设输入的表达式只会出现加（“+”），减 “-”， 乘“*”，除 “/” 四个运算符, 表达式中只会出现正整数

2. 输入：“$34+13*9 + 44-12/3$”  输出：191

idea：

### 算法设计思路

该算法使用经典的双栈方法来计算算术表达式的值，一个栈用于存放操作数（`values`），另一个栈用于存放运算符（`ops`）。

1.  **初始化**：
    *   创建两个空栈：`values` (操作数栈) 和 `ops` (运算符栈)。
    *   定义一个函数 `precedence(op)` 来返回运算符的优先级（例如，`*` 和 `/` 的优先级高于 `+` 和 `-`）。
    *   定义一个函数 `apply_op()`，它会：
        *   从 `ops` 栈弹出一个运算符。
        *   从 `values` 栈弹出两个操作数（注意顺序，通常是先弹出的为右操作数，后弹出的为左操作数）。
        *   执行运算。
        *   将运算结果压回 `values` 栈。

2.  **遍历表达式**：
    *   从左到右逐个处理输入表达式字符串中的字符。
    *   **处理空格**：如果当前字符是空格，则跳过。
    *   **处理数字**：如果当前字符是数字，则继续读取后续的数字字符，以形成一个完整的（可能多位的）数字。将这个数字转换为整数后压入 `values` 栈。
    *   **处理运算符**：如果当前字符是一个运算符 (`+`, `-`, `*`, `/`)：
        *   当 `ops` 栈不为空，且栈顶运算符的优先级 **大于或等于** 当前扫描到的运算符的优先级时，重复执行 `apply_op()`。这个步骤确保了高优先级的运算符或者相同优先级下左边的运算符先被计算。
        *   完成上述循环后（或如果循环条件不满足），将当前扫描到的运算符压入 `ops` 栈。

3.  **处理剩余运算**：
    *   当整个表达式字符串都扫描完毕后，`ops` 栈中可能还剩余一些运算符。
    *   重复执行 `apply_op()` 直到 `ops` 栈为空。

4.  **获取结果**：
    *   此时，`values` 栈中应该只剩下一个数字，这个数字就是整个表达式的计算结果。

**运算符优先级**：
*   `*`, `/`：优先级 2
*   `+`, `-`：优先级 1

**注意事项**：
*   题目规定表达式中只会出现正整数和四种基本运算符。
*   除法运算应按整数除法处理，例如 `12/3` 结果为 `4`。

### 时间复杂度分析

*   算法需要从左到右遍历输入表达式字符串一次。设字符串长度为 N。
*   每个数字（操作数）被解析并压入 `values` 栈一次，之后在参与运算时被弹出一次。
*   每个运算符被压入 `ops` 栈一次，并在其优先级允许或在表达式末尾时被弹出并用于运算一次。
*   数字的解析（处理多位数字）总体上与数字的位数成正比，所有数字的总位数是 O(N)。
*   `apply_op` 函数执行固定数量的操作。
*   在处理运算符时，内部的 `while` 循环（用于 `apply_op`）虽然看起来可能多次执行，但每个运算符最多被压入和弹出一遍。因此，所有 `apply_op` 的总调用次数受限于运算符的数量。
*   综合来看，每个字符、数字和运算符都只被有效处理常数次。
*   因此，总的时间复杂度为 **O(N)**，其中 N 是输入表达式字符串的长度。

### 空间复杂度分析

*   `values` 栈：在最坏的情况下（例如，表达式 "1+2+3+...+k"），栈中可能存储 O(M) 个数字，其中 M 是表达式中数字的数量。
*   `ops` 栈：在最坏的情况下（例如，表达式 "1+2\*3\*4\*..."，优先级递增的运算符序列），栈中可能存储 O(K) 个运算符，其中 K 是表达式中运算符的数量。
*   M 和 K 都与输入字符串的长度 N 相关。
*   因此，空间复杂度为 **O(N)**。

````python
# add your idea here
# your algorithm time complexity is: 
def calculate_expression(expression_str: str) -> int:
    """
    根据双栈原理计算算术表达式的值。
    表达式只包含正整数、+、-、*、/ 和空格。
    """

    def precedence(op: str) -> int:
        """返回运算符的优先级"""
        if op == '+' or op == '-':
            return 1
        if op == '*' or op == '/':
            return 2
        return 0  # 通常用于处理括号等，此处用作默认低优先级

    def apply_op(ops_stack: list, values_stack: list):
        """从栈中取出运算符和操作数进行计算，并将结果压回操作数栈"""
        try:
            op = ops_stack.pop()
            right_operand = values_stack.pop()
            left_operand = values_stack.pop()

            if op == '+':
                values_stack.append(left_operand + right_operand)
            elif op == '-':
                values_stack.append(left_operand - right_operand)
            elif op == '*':
                values_stack.append(left_operand * right_operand)
            elif op == '/':
                if right_operand == 0:
                    raise ZeroDivisionError("Cannot divide by zero")
                # 题目示例暗示整数除法
                values_stack.append(left_operand // right_operand)
        except IndexError:
            raise ValueError("Invalid expression format or insufficient operands/operators")

    values_stack = []  # 操作数栈
    ops_stack = []     # 运算符栈
    
    i = 0
    n = len(expression_str)
    
    while i < n:
        char = expression_str[i]
        
        if char.isspace():
            i += 1
            continue
            
        elif char.isdigit():
            num_str = ""
            # 读取完整的数字
            while i < n and expression_str[i].isdigit():
                num_str += expression_str[i]
                i += 1
            values_stack.append(int(num_str))
            # i 已经指向数字后的字符，所以用 continue 开始下一次主循环
            continue 
            
        elif char in ['+', '-', '*', '/']:
            # 当ops栈非空，且栈顶运算符优先级 >= 当前运算符优先级
            while ops_stack and precedence(ops_stack[-1]) >= precedence(char):
                apply_op(ops_stack, values_stack)
            ops_stack.append(char)
            i += 1
        else:
            # 处理无效字符，如果需要
            raise ValueError(f"Invalid character in expression: {char}")
            # i += 1 # or skip

    # 处理栈中剩余的所有运算符
    while ops_stack:
        apply_op(ops_stack, values_stack)
        
    if len(values_stack) == 1:
        return values_stack[0]
    elif not values_stack and not expression_str.strip(): # Handle empty or whitespace-only string
        raise ValueError("Expression is empty")
    else:
        raise ValueError("Invalid expression leading to improper stack state at the end")

 
print(f"结果: {calculate_expression('3+5 * 8 -6')}") # 期望输出: 37
print(f"结果: {calculate_expression('34+13*9 + 44-12/3')}") # 期望输出: 191
print(f"结果: {calculate_expression('100')}") # 期望输出: 100
print(f"结果: {calculate_expression(' 2 * 3 + 5 / 6 * 3 + 15')}") # 2*3 + 0*3 + 15 = 6 + 0 + 15 = 21

# your algorithm time complexity is: O(N)
````

## 问题 4:  

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]



### 算法设计思路

这个问题可以使用栈（Stack）数据结构来高效地模拟星球碰撞的过程。栈的特性“后进先出”（LIFO）非常适合处理这种一维直线上的碰撞，特别是当一个星球向左移动并可能与之前路径上所有向右移动的星球发生交互时。

1.  **初始化**：
    *   创建一个空栈 `stack`。这个栈将用来存储那些到目前为止还存活并且正在向右移动的星球，或者是那些已经处理过碰撞且没有被摧毁的向左移动的星球。

2.  **遍历星球**：
    *   从左到右遍历输入的星球数组 `A`。对于每个星球 `current_planet`：
        *   **Case 1: `current_planet` 向右移动 (`current_planet > 0`)**
            *   直接将 `current_planet` 压入 `stack`。这些向右移动的星球暂时不会与任何东西碰撞（除非后续有向左移动的星球）。

        *   **Case 2: `current_planet` 向左移动 (`current_planet < 0`)**
            *   这是一个关键的碰撞处理阶段。当一个向左移动的星球出现时，它可能会与栈顶（即最右边的、之前遇到的、向右移动的）星球发生碰撞。
            *   我们需要一个循环来处理 `current_planet` 与栈中所有可能碰撞的星球的交互：
                *   **循环条件**：栈不为空 (`stack` is not empty) **且** 栈顶星球向右移动 (`stack.top() > 0`) **且** `current_planet` 仍然存活（因为 `current_planet` 可能在某次碰撞中被摧毁）。
                *   **获取栈顶星球**：`right_moving_planet = stack.top()`。
                *   **比较质量**（绝对值）：
                    *   **如果 `abs(current_planet) > abs(right_moving_planet)`**：
                        *   向左的星球 `current_planet` 质量更大，所以栈顶的向右星球 `right_moving_planet` 被摧毁。
                        *   将 `right_moving_planet` 从栈中弹出 (`stack.pop()`)。
                        *   `current_planet` 继续存活并可能与栈中下一个向右移动的星球碰撞（循环继续）。
                    *   **如果 `abs(current_planet) < abs(right_moving_planet)`**：
                        *   向右的星球 `right_moving_planet` 质量更大，所以向左的星球 `current_planet` 被摧毁。
                        *   `current_planet` 的旅程结束，跳出内部的碰撞处理循环（例如，可以将 `current_planet` 标记为已摧毁或直接 `break`）。栈顶的 `right_moving_planet` 保持不变。
                    *   **如果 `abs(current_planet) == abs(right_moving_planet)`**：
                        *   根据题目规则：“质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变”。
                        *   这意味着向左的 `current_planet` 被摧毁。
                        *   栈顶的向右星球 `right_moving_planet` 保持不变（不弹出）。
                        *   `current_planet` 的旅程结束，跳出内部的碰撞处理循环。
            *   **循环结束后**：
                *   如果 `current_planet` 在所有碰撞后仍然存活（即它没有被任何向右移动的星球摧毁），那么它将继续向左移动，并且不会再与之前已经处理过的、在它左边的星球发生碰撞。此时，将存活的 `current_planet` 压入 `stack`。
                *   注意：如果 `current_planet` 在碰撞中被摧毁，则它不应被压入栈。

3.  **结果**：
    *   当遍历完所有输入星球后，`stack` 中剩余的星球就是最终存续下来的星球。栈中的元素顺序即为它们在直线上的相对顺序。
    *   将栈中的元素转换为数组并返回。

### 时间复杂度分析

*   每个星球 `p` 从输入数组 `A` 中被处理一次。
*   当星球 `p` 被压入栈时，这是 O(1) 操作。
*   一个星球最多被压入栈一次。
*   当一个向左移动的星球 `current_planet` 出现时，它会与栈顶的向右移动的星球进行比较。
    *   如果 `current_planet` 摧毁了栈顶星球，栈顶星球被弹出。一个星球最多被弹出一次。
    *   如果 `current_planet` 被摧毁，它停止。
*   每个星球要么最终保留在栈中，要么在碰撞中被摧毁并从栈中移除（如果是向右的）或不加入栈中（如果是向左的且被摧毁）。
*   因此，每个星球参与的压栈、弹栈和比较操作的总次数是有限的（每个星球最多入栈一次，出栈一次）。
*   所以，总的时间复杂度为 **O(N)**，其中 N 是输入星球的数量。

### 空间复杂度分析

*   在最坏的情况下，如果所有星球都向同一个方向移动（例如，所有都向右，或所有都向左且没有碰撞），栈中可能会存储所有 N 个星球。
*   因此，空间复杂度为 **O(N)**。

````python
from typing import List

def planet_collision(A: List[int]) -> List[int]:
    """
    模拟星球碰撞，返回最终存续的星球情况。
    """
    stack = [] # 用列表模拟栈

    for planet in A:
        if planet > 0:
            # 星球向右移动，直接入栈
            stack.append(planet)
        else:
            # 星球向左移动，处理碰撞
            planet_alive = True
            while stack and stack[-1] > 0 and planet_alive:
                # 栈顶是向右移动的星球
                right_moving_planet = stack[-1]
                
                if abs(planet) > abs(right_moving_planet):
                    # 向左的星球质量大，摧毁向右的星球
                    stack.pop()
                    # 向左的星球继续（planet_alive 保持 True）
                elif abs(planet) < abs(right_moving_planet):
                    # 向右的星球质量大，向左的星球被摧毁
                    planet_alive = False # 当前向左的星球不再存活
                    # 向右的星球保持在栈顶
                else: # abs(planet) == abs(right_moving_planet)
                    # 质量相同，向左的星球消失，向右的保持
                    planet_alive = False # 当前向左的星球不再存活
                    # 向右的星球保持在栈顶
            
            if planet_alive:
                # 如果向左的星球在与所有向右的星球碰撞后仍然存活
                # (或者一开始栈就空了，或者栈顶就是向左的)
                # 那么它就加入到最终的星球序列中
                stack.append(planet)
                
    return stack

# 示例测试
A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
output1 = planet_collision(A1)
print(f"输入 A={A1}, 输出: {output1}")

A2 = [23, -8, 9, -3, -7, 9, -23, 22]
output2 = planet_collision(A2)
print(f"输入 A={A2}, 输出: {output2}")

A3 = [5, 10, -5]
output3 = planet_collision(A3)
print(f"输入 A={A3}, 输出: {output3}")

A4 = [10, 2, -5]
output4 = planet_collision(A4)
print(f"输入 A={A4}, 输出: {output4}")

A5 = [-1, -2, -3]
output5 = planet_collision(A5)
print(f"输入 A={A5}, 输出: {output5}")

A6 = [1, 2, 3]
output6 = planet_collision(A6)
print(f"输入 A={A6}, 输出: {output6}")

# 算法时间复杂度已在Markdown中分析为 O(N)
# 算法空间复杂度已在Markdown中分析为 O(N)
````

使用python编程完成上述任务，并计算时间复杂度


````python
# add your idea here
class TreeNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def insert(root: TreeNode, key: int) -> TreeNode:
    """将元素插入二叉搜索树"""
    if root is None:
        return TreeNode(key)
    
    if key < root.key:
        root.left = insert(root.left, key)
    elif key > root.key: # 标准BST不插入重复键，如果需要处理重复，策略会不同
        root.right = insert(root.right, key)
    # If key == root.key, we can choose to do nothing (don't insert duplicates)
    # or handle it based on specific requirements (e.g., store count).
    # For this problem, simply not inserting duplicates is fine.
    return root

def find_min_value_node(node: TreeNode) -> TreeNode:
    """找到给定子树中的最小值的节点 (最左边的节点)"""
    current = node
    while current and current.left is not None:
        current = current.left
    return current

def delete_node(root: TreeNode, key: int) -> TreeNode:
    """从二叉搜索树中删除指定键的节点"""
    if root is None:
        return root # 键不存在

    # 递归查找要删除的节点
    if key < root.key:
        root.left = delete_node(root.left, key)
    elif key > root.key:
        root.right = delete_node(root.right, key)
    else:
        # 找到了要删除的节点 (root.key == key)
        # Case 1: 节点是叶子节点 (没有子节点) 或只有一个子节点
        if root.left is None:
            temp = root.right
            root = None # 在Python中，这只是解除了局部变量root对节点的引用
                        # 真正的删除依赖于调用者重新分配root.left或root.right
            return temp # 返回右子节点（或None，如果也是None）
        elif root.right is None:
            temp = root.left
            root = None
            return temp # 返回左子节点

        # Case 2: 节点有两个子节点
        # 找到右子树中的最小节点 (中序后继)
        temp = find_min_value_node(root.right)
        
        # 将中序后继的值复制到当前节点
        root.key = temp.key
        
        # 从右子树中删除这个中序后继节点
        root.right = delete_node(root.right, temp.key)
        
    return root

def in_order_traversal(root: TreeNode, result_list: list):
    """中序遍历二叉搜索树，并将结果存入列表"""
    if root:
        in_order_traversal(root.left, result_list)
        result_list.append(root.key)
        in_order_traversal(root.right, result_list)

def process_bst_operations(nums: list, key_to_delete: int) -> list:
    """
    构建BST，删除节点，然后中序遍历。
    """
    root = None
    # 1. 建立二叉搜索树
    # 使用set来处理输入数组中的重复元素，确保BST中只有唯一值
    unique_nums = sorted(list(set(nums))) # 可选：对唯一值排序再插入可能产生更平衡的树
                                        # 但题目通常是按原数组顺序插入
    
    # 按数组原顺序（去重后）插入
    processed_for_bst = []
    seen = set()
    for num in nums:
        if num not in seen:
            processed_for_bst.append(num)
            seen.add(num)

    for num in processed_for_bst: # 使用处理过去重并保持相对顺序的列表
        root = insert(root, num)
        
    # 2. 删除二叉树中的元素
    root = delete_node(root, key_to_delete)
    
    # 3. 中序遍历输出
    result = []
    in_order_traversal(root, result)
    return result

# 示例输入
nums_example = [9, -3, -10, 0, 9, 7, 33]
key_to_delete_example = 0

# 执行操作
output_elements = process_bst_operations(nums_example, key_to_delete_example)
print(f"输入数组: {nums_example}")
print(f"删除元素: {key_to_delete_example}")
print(f"中序遍历结果: {output_elements}") # 期望: 根据插入顺序和删除逻辑

# 另一组测试，确保删除根节点等情况
nums_example2 = [50, 30, 70, 20, 40, 60, 80]
# BST:
#       50
#      /  \
#    30    70
#   / \   / \
# 20  40 60  80
key_to_delete2 = 50 # 删除根
output2 = process_bst_operations(nums_example2, key_to_delete2)
# 期望中序遍历 (例如，用60替换50): [20, 30, 40, 60, 70, 80]
print(f"\n输入数组: {nums_example2}, 删除元素: {key_to_delete2}, 中序遍历结果: {output2}")

key_to_delete3 = 20 # 删除叶子
output3 = process_bst_operations(nums_example2, key_to_delete3)
print(f"输入数组: {nums_example2}, 删除元素: {key_to_delete3}, 中序遍历结果: {output3}")


# 时间复杂度分析:
# - N 是输入数组中唯一元素的数量。
# - 构建BST (N次插入):
#   - 平均情况 (树相对平衡): O(N log N)，因为每次插入平均耗时 O(log H)，H 为树高，平均 H ~ log N。
#   - 最坏情况 (树极度不平衡，如链状): O(N^2)，因为每次插入最坏耗时 O(H)，H ~ N。
#     (对于题目给的无序数组，通常更接近平均情况，但最坏情况仍有可能)
# - 删除一个元素:
#   - 查找元素 + 实际删除（可能涉及查找中序后继）:
#   - 平均情况: O(log N)
#   - 最坏情况: O(N)
# - 中序遍历: O(N)，因为每个节点访问一次。
#
# 总体时间复杂度:
# - 平均情况: O(N log N) (主要由BST构建主导)
# - 最坏情况: O(N^2) (主要由BST构建主导)
#
# 对于这个问题，由于输入是 "无序数组"，可以假设平均情况的性能更有代表性，但严格来说，
# 最坏情况下的时间复杂度是 O(N^2)。
#
# 空间复杂度:
# - O(N) 用于存储BST本身。
# - O(H) 用于递归栈，H是树的高度。最坏情况 H=N (O(N))，平均情况 H=logN (O(log N))。
# 所以空间复杂度最坏是 O(N)。
````

## 问题 6  

> 给定一个包含大写字母和小写字母的字符串 s ，返回 字符串包含的 **最长的回文子串的长度** 。请注意 区分大小写 。比如 "Aa" 不能当做一个回文字符串。
>

请设计一个算法解决上述问题，只需要输出最长回文子串的长度，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g. 输入： s="adccaccd"，  输出：7。 最长回文子串为："dccaccd", 长度为7


idea：

````python
# add your idea here
# add your idea here
def longest_palindrome_substring_length(s: str) -> int:
    """
    返回字符串 s 中最长回文子串的长度。
    区分大小写。
    """
    n = len(s)
    if n < 1: # 处理空串或长度为1的串
        return n 
    
    max_len = 1 # 任何非空字符串至少有长度为1的回文 (单个字符)

    def expand_around_center(left: int, right: int) -> int:
        """
        从给定的中心 (left, right) 向两边扩展，返回找到的回文串的长度。
        """
        l, r = left, right
        while l >= 0 and r < n and s[l] == s[r]:
            l -= 1
            r += 1
        # 循环结束后，s[l+1...r-1] 是以 (left, right) 为中心的最长回文
        # 其长度是 (r-1) - (l+1) + 1 = r - l - 1
        return r - l - 1

    for i in range(n):
        # 奇数长度回文串，中心是 s[i]
        # 初始中心 (i, i)
        len1 = expand_around_center(i, i)
        if len1 > max_len:
            max_len = len1
            
        # 偶数长度回文串，中心是 s[i] 和 s[i+1] 之间
        # 初始中心 (i, i+1)
        # 确保 i+1 不越界
        if i + 1 < n:
            len2 = expand_around_center(i, i + 1)
            if len2 > max_len:
                max_len = len2
                
    return max_len

# 示例测试
s1 = "adccaccd"
output1 = longest_palindrome_substring_length(s1)
print(f"输入: s=\"{s1}\", 输出: {output1}")

s2 = "babad"
output2 = longest_palindrome_substring_length(s2)
print(f"输入: s=\"{s2}\", 输出: {output2}")

s3 = "cbbd"
output3 = longest_palindrome_substring_length(s3)
print(f"输入: s=\"{s3}\", 输出: {output3}")

s4 = "a"
output4 = longest_palindrome_substring_length(s4)
print(f"输入: s=\"{s4}\", 输出: {output4}")

s5 = ""
output5 = longest_palindrome_substring_length(s5)
print(f"输入: s=\"{s5}\", 输出: {output5}")

s6 = "Aa"
output6 = longest_palindrome_substring_length(s6)
print(f"输入: s=\"{s6}\", 输出: {output6}")

s7 = "racecar"
output7 = longest_palindrome_substring_length(s7)
print(f"输入: s=\"{s7}\", 输出: {output7}")


# your algorithm time complexity is: O(n^2)
# your algorithm space complexity is: O(1)

````

输入： [1, 5, 12, 33, 34,35]  输出：基站数目为3， 基站位置为[1，12，33]



idea：

````python
from typing import List, Tuple

def place_base_stations(houses: List[int]) -> Tuple[int, List[int]]:
    """
    计算覆盖所有房子所需的最少基站数量及其位置。
    每个基站覆盖其位置左右4公里范围。
    """
    n = len(houses)
    if n == 0:
        return 0, []

    station_positions: List[int] = []
    num_stations = 0
    i = 0
    while i < n:
        num_stations += 1
        # houses[i] 是第一个需要被这个新基站覆盖的房子。
        # 为了最大化此基站的向右覆盖范围，将其放置在 houses[i] + 4。
        current_station_pos = houses[i] + 4
        station_positions.append(current_station_pos)
        
        # 此基站覆盖的范围是 [houses[i], houses[i] + 8]。
        # 所以，它能覆盖到的最远的房子坐标是 houses[i] + 8。
        coverage_limit = houses[i] + 8
        
        # 找到下一个未被覆盖的房子
        current_house_idx_covered_by_this_station = i
        while current_house_idx_covered_by_this_station < n and \
              houses[current_house_idx_covered_by_this_station] <= coverage_limit:
            current_house_idx_covered_by_this_station += 1
        
        # 更新i为下一个需要覆盖的房子的索引
        i = current_house_idx_covered_by_this_station
        
    return num_stations, station_positions

# 示例测试
houses_example = [1, 5, 12, 33, 34, 35]
num, positions = place_base_stations(houses_example)
print(f"输入: {houses_example}")
print(f"基站数目为{num}，基站位置为{positions}")
# 期望输出 (对于此贪心策略): 基站数目为3，基站位置为[5, 16, 37]
# 示例中给出的位置是 [1, 12, 33]，这也是一个有效的3基站方案。
# 贪心算法保证了最少的基站数目，但具体位置可能有多种最优解。

houses_example_2 = [0, 1, 2, 8, 9, 10, 15, 16, 17, 25]
num_2, positions_2 = place_base_stations(houses_example_2)
print(f"\n输入: {houses_example_2}")
print(f"基站数目为{num_2}，基站位置为{positions_2}")
# 0 -> station at 4, covers [0,8]. Next house is 9.
# 9 -> station at 13, covers [9,17]. Next house is 25.
# 25 -> station at 29, covers [25,33].
# 期望输出: 基站数目为3，基站位置为[4, 13, 29]

# your algorithm time complexity is: O(N)
````

#### 证明算法（策略：放置基站于 `houses[i] + 4`）产生最优解决方案

我们将使用“贪心选择 stays ahead”的证明方法。该方法表明，贪心算法在每一步都做出一个选择，这个选择至少与某个最优解中的选择一样“好”，并且能够导向一个全局最优解。

**定义**：
*   设 `H = {h_0, h_1, ..., h_{n-1}}` 为按坐标排序的房子集合。
*   贪心算法 G 选择的基站位置序列为 `G_S = {g_1, g_2, ..., g_k}`。
*   一个最优算法 OPT 选择的基站位置序列为 `O_S = {o_1, o_2, ..., o_m}`，其中 `m` 是可能的最小基站数。我们希望证明 `k = m`。
*   对于一个基站放置在 `p`，其覆盖的最右点为 `p + 4`。但是，为了论证，我们更关心它能覆盖到的房子区域。一个基站 `g` 放置在 `h + 4` (其中 `h` 是它覆盖的第一个房子) 能覆盖所有房子直到 `h + 8`。我们称 `R(g) = h + 8` 为该基站的有效右覆盖边界。

**贪心选择性质**：
当贪心算法需要放置一个新基站时，它选择第一个未被覆盖的房子 `h_f`，然后将基站 `g` 放置在 `h_f + 4`。这个基站 `g` 覆盖的范围是 `[h_f, h_f + 8]`。

**证明**：
我们通过归纳法证明，对于每个 `j ≥ 1`，如果 `g_j` 是贪心算法放置的第 `j` 个基站，`o_j` 是某个最优解中（按位置或覆盖范围排序的）第 `j` 个基站，那么 `R(g_j) ≥ R(o_j)`。这意味着第 `j` 个贪心基站的有效右覆盖边界至少与第 `j` 个最优基站的有效右覆盖边界一样远。

**基础情况 (j=1)**：
*   贪心算法 G 选择第一个房子 `h_0`。它放置第一个基站 `g_1` 在 `h_0 + 4`。所以 `R(g_1) = h_0 + 8`。
*   任何最优解 OPT 必须覆盖 `h_0`。设 `o_1` 是 OPT 中覆盖 `h_0` 的基站（如果 OPT 中有多个基站覆盖 `h_0`，可以考虑最左边的那个，或者对结果影响最大的那个）。为了覆盖 `h_0`，`o_1` 必须放置在 `p_1` 使得 `p_1 - 4 ≤ h_0`。为了最大化其右覆盖，OPT 会选择最大的 `p_1`，即 `p_1 = h_0 + 4`。此时，`R(o_1) = (h_0 + 4) + 4 = h_0 + 8`。
*   如果 OPT 选择的 `p_1 < h_0 + 4`，则 `R(o_1) = p_1 + 4 < h_0 + 8`。
*   因此，在所有情况下，`R(g_1) ≥ R(o_1)`。贪心算法的第一个基站覆盖的区域至少和最优解的第一个基站一样远。

**归纳步骤**：
*   假设对于某个 `j ≥ 1`，我们已经证明了 `R(g_j) ≥ R(o_j)`。
*   现在考虑放置第 `j+1` 个基站。
*   设 `h_fG` 是在放置了 `g_1, ..., g_j` 之后第一个未被覆盖的房子。贪心算法 G 会放置 `g_{j+1}` 在 `h_fG + 4`，因此 `R(g_{j+1}) = h_fG + 8`。
*   设 `h_fO` 是在放置了 `o_1, ..., o_j` 之后第一个未被覆盖的房子。最优解 OPT 必须放置 `o_{j+1}` 来覆盖 `h_fO`。`o_{j+1}` 会被放置在某个位置 `p_{j+1}` 使得 `p_{j+1} - 4 ≤ h_fO`，并且其有效右覆盖为 `R(o_{j+1}) = p_{j+1} + 4`。为了最大化这个覆盖，`p_{j+1}` 最多为 `h_fO + 4`，所以 `R(o_{j+1}) ≤ h_fO + 8`。
*   根据归纳假设 `R(g_j) ≥ R(o_j)`，这意味着由 `g_1, ..., g_j` 覆盖的所有房子也一定被 `o_1, ..., o_j` 覆盖（因为 `g` 的覆盖延伸得更远或一样远）。所以，第一个未被 `g_1, ..., g_j` 覆盖的房子 `h_fG` 必然在第一个未被 `o_1, ..., o_j` 覆盖的房子 `h_fO` 的右边或同一位置，即 `h_fG ≥ h_fO`。
*   因此，`R(g_{j+1}) = h_fG + 8 ≥ h_fO + 8 ≥ R(o_{j+1})`。
*   归纳成立。

**结论**：
由于 `R(g_j) ≥ R(o_j)` 对所有 `j` 成立，如果最优解 OPT 使用 `m` 个基站来覆盖所有房子，那么 `R(o_m)` 必须覆盖到最后一个房子 `h_{n-1}`。因为 `R(g_m) ≥ R(o_m)`，所以贪心算法的第 `m` 个基站 `g_m` 也肯定覆盖了所有房子。这意味着贪心算法 G 最多使用 `m` 个基站。由于 `m` 是最优（最小）数量，贪心算法使用的基站数量 `k` 必须等于 `m`。
因此，贪心算法产生了一个最优解决方案，使用了最少数量的基站。

## 问题 8  

> 给定由n个正整数组成的一个集合$S = \{a_1, a_2，···，a_n\}$和一个正整数W，设计一个算法确定是否存在S的一个子集 $K \subseteq S$, 使K中所有数之和为 $W$, 如果存在返回“True”，否则返回“False”

请设计一个时间复杂度为$O(nW)$动态规划算法，解决上述问题，分析算法的设计思路，并且基于python编程实现（不需要输出子集）。

e.g. 

输入：S = {1,4,7,3,5}， W = 11，输出：True。   因为K可以是{4,7}。



idea：

````python
from typing import List, Set # Though Set is not strictly needed for the input type hint if it's a list

def subset_sum_dp(S: List[int], W: int) -> bool:
    """
    确定是否存在S的一个子集K，使K中所有数之和为W。
    使用动态规划实现。
    S: 由n个正整数组成的集合 (以列表形式给出)
    W: 正整数目标和
    """
    n = len(S)
    
    # dp[i][j] 为 True 如果和 j 可以用 S 的前 i 个元素得到
    # dp 表的大小是 (n+1) x (W+1)
    # 行对应元素数量 (0 to n), 列对应和 (0 to W)
    dp = [[False for _ in range(W + 1)] for _ in range(n + 1)]

    # 初始化
    # 如果和是0，则答案是 True (空子集)
    for i in range(n + 1):
        dp[i][0] = True
        
    # 如果元素数量是0 (空集 S)，但和不是0，则答案是 False
    # (dp[0][1...W] 已经是 False by default)

    # 填充dp表
    for i in range(1, n + 1): # 对应 S 中的第 i 个元素 S[i-1]
        current_element = S[i-1]
        for j in range(1, W + 1): # 目标和 j
            # 情况1: 不包括当前元素 S[i-1]
            # 那么结果取决于是否可以用前 i-1 个元素达到和 j
            dp[i][j] = dp[i-1][j]
            
            # 情况2: 包括当前元素 S[i-1]
            # 这只有在 j >= S[i-1] 时才可能
            if current_element <= j:
                # 如果包括 S[i-1]，那么我们需要检查是否可以用前 i-1 个元素达到和 (j - S[i-1])
                # dp[i][j] 是 (不包括S[i-1]的结果) OR (包括S[i-1]的结果)
                dp[i][j] = dp[i][j] or dp[i-1][j - current_element]
                
    return dp[n][W]

# 示例测试
S_example = {1, 4, 7, 3, 5} # 题目用集合表示，但通常作为列表输入给函数
W_example = 11
# 将集合转换为列表以便索引，或者直接在函数内处理
output_example = subset_sum_dp(list(S_example), W_example)
print(f"输入: S = {S_example}， W = {W_example}，输出: {output_example}") # 期望输出: True

S_example2 = [2, 3, 7, 8, 10]
W_example2 = 11
output_example2 = subset_sum_dp(S_example2, W_example2)
print(f"输入: S = {S_example2}， W = {W_example2}，输出: {output_example2}") # 期望输出: True (3+8)

S_example3 = [3, 34, 4, 12, 5, 2]
W_example3 = 30
output_example3 = subset_sum_dp(S_example3, W_example3)
print(f"输入: S = {S_example3}， W = {W_example3}，输出: {output_example3}") # 期望输出: False

S_example4 = [1, 2, 3]
W_example4 = 7
output_example4 = subset_sum_dp(S_example4, W_example4)
print(f"输入: S = {S_example4}， W = {W_example4}，输出: {output_example4}") # 期望输出: False (1+2+3=6)

# your algorithm time complexity is: O(nW)
````

> 给定一个n个物品的集合。物体的重量为$w_1, w_2，…、w_n$，物品的价值分别是$v_1、v_2、…v_n$。给你**两个**重量为 $c$ 的背包。如果你带了一个东西，它可以放在一个背包里，也可以放在另一个背包里，但不能同时放在两个背包里。所有权重和价值都是正整数。

1. 设计一个时间复杂度为 $O(nc^2)$ 的动态规划算法，确定可以放入两个背包的物体的最大价值。分析算法设计思路，并基于python编程实现
2. \* 修改1中的算法，输出每个背包的内容（物品对应下标）。

e.g.: 

输入 V=[1,3,2,5,8,7], W=[1,3,2,5,8,7], c=7, 输出：最大价值=14，背包装的物品为：[6] [4，3] （同一个背包中物品装入顺序对结果无影响）  


idea：

### 算法设计思路

这是一个扩展的0/1背包问题，因为我们有两个容量相同的背包。我们需要最大化放入这两个背包的物品的总价值。

**1. 定义 DP 状态**

我们使用一个三维 DP 数组 `dp[i][j][k]`，其含义如下：
`dp[i][j][k]` = 考虑前 `i` 个物品（从物品0到物品 `i-1`），第一个背包已用容量为 `j`，第二个背包已用容量为 `k` 时，能够获得的最大总价值。

*   `i`：物品的索引，从 `0` 到 `n` ( `n` 是物品总数)。`i` 表示考虑了前 `i` 个物品。
*   `j`：第一个背包的当前已用重量，从 `0` 到 `c` (背包容量)。
*   `k`：第二个背包的当前已用重量，从 `0` 到 `c` (背包容量)。

DP 表的维度将是 `(n + 1) x (c + 1) x (c + 1)`。

**2. 初始化 DP 表**

*   `dp[0][0][0] = 0`：不考虑任何物品，两个背包都为空，总价值为0。
*   所有其他的 `dp[0][j][k]` 可以初始化为一个非常小的值（例如负无穷），表示这些状态不可达或无效。或者，在填充过程中，如果某个状态的前置状态是无效的，那么当前状态也认为是无效的，除非它能通过当前物品直接达到。对于0/1背包，通常将无法达到的状态的价值视为0（如果不选任何物品），或者在计算时通过判断前置状态是否有效来处理。
    为了简化，我们可以将所有 `dp` 值初始化为0，因为价值是正整数。

**3. 填充 DP 表（递推关系）**

我们遍历每个物品 `i` (从1到 `n`，对应物品 `items[i-1]`)，然后遍历第一个背包的所有可能容量 `j` (从0到 `c`)，再遍历第二个背包的所有可能容量 `k` (从0到 `c`)。

对于当前物品 `item = items[i-1]` (其重量为 `w_item = W[i-1]`，价值为 `v_item = V[i-1]`)，我们有以下几种选择：

*   **选择1: 不将物品 `item` 放入任何背包。**
    这种情况下，最大价值与不考虑此物品，且背包容量保持为 `j` 和 `k` 时的最大价值相同。
    `value1 = dp[i-1][j][k]`

*   **选择2: 将物品 `item` 放入第一个背包。**
    这只有在第一个背包的当前容量 `j` 能够容纳 `item` (`j >= w_item`) 时才可能。
    如果可以放入，那么最大价值是 `item` 的价值加上不考虑此物品，第一个背包容量为 `j - w_item`，第二个背包容量为 `k` 时的最大价值。
    `value2 = dp[i-1][j - w_item][k] + v_item` (如果 `j >= w_item`，否则此选项无效，可视为负无穷或不考虑)

*   **选择3: 将物品 `item` 放入第二个背包。**
    这只有在第二个背包的当前容量 `k` 能够容纳 `item` (`k >= w_item`) 时才可能。
    如果可以放入，那么最大价值是 `item` 的价值加上不考虑此物品，第一个背包容量为 `j`，第二个背包容量为 `k - w_item` 时的最大价值。
    `value3 = dp[i-1][j][k - w_item] + v_item` (如果 `k >= w_item`，否则此选项无效)

因此，`dp[i][j][k]` 的值是这三种选择中能得到的最大价值：
`dp[i][j][k] = max(value1, value2_if_possible, value3_if_possible)`

具体地：
`dp[i][j][k] = dp[i-1][j][k]` (基本情况：不选当前物品)
If `j >= W[i-1]`:
  `dp[i][j][k] = max(dp[i][j][k], dp[i-1][j - W[i-1]][k] + V[i-1])`
If `k >= W[i-1]`:
  `dp[i][j][k] = max(dp[i][j][k], dp[i-1][j][k - W[i-1]] + V[i-1])`

**4. 最终结果**

在填充完整个 `dp` 表后，考虑了所有 `n` 个物品。此时，我们需要找到 `dp[n][j][k]` 在所有可能的 `0 <= j <= c` 和 `0 <= k <= c` 中的最大值。这个最大值就是所求的最大总价值。
`max_total_value = max(dp[n][j][k])` for all `0 <= j <= c`, `0 <= k <= c`.

**时间复杂度分析**
*   我们有 `n` 个物品。
*   对于每个物品，我们需要填充 `(c+1) x (c+1)` 个状态。
*   每个状态的计算是常数时间 (几次比较和加法)。
*   因此，总的时间复杂度是 **O(n * c^2)**。

**空间复杂度分析**
*   我们使用了一个大小为 `(n+1) x (c+1) x (c+1)` 的 DP 表。
*   因此，空间复杂度为 **O(n * c^2)**。
    (可以通过状态压缩将空间优化到 O(c^2)，只保留当前和前一个物品的状态，但这会使路径回溯更复杂。)

---

**第2部分 (\*修改算法输出每个背包的内容)**

要回溯并找到每个背包中的物品，我们需要在填充DP表后，从 `dp[n][j_final][k_final]` 开始（其中 `j_final` 和 `k_final` 是得到最大总价值时两个背包的容量），反向推导每个物品的选择。

1.  **初始化**：
    *   `bag1_items = []`, `bag2_items = []`
    *   `current_value = max_total_value`
    *   `current_j = j_final`, `current_k = k_final`

2.  **从最后一个物品开始回溯** (从 `i = n` 向下到 `1`)：
    *   获取当前物品 `item = items[i-1]` (重量 `w_item`, 价值 `v_item`)。
    *   **检查是否放入了第二个背包**：
        如果 `current_k >= w_item` 且 `current_value == dp[i-1][current_j][current_k - w_item] + v_item`：
        这意味着物品 `item` 被放入了第二个背包。
        将其添加到 `bag2_items`。
        更新 `current_value -= v_item`。
        更新 `current_k -= w_item`。
        继续下一个物品 `i-1`。
    *   **否则，检查是否放入了第一个背包**：
        如果 `current_j >= w_item` 且 `current_value == dp[i-1][current_j - w_item][current_k] + v_item`：
        这意味着物品 `item` 被放入了第一个背包。
        将其添加到 `bag1_items`。
        更新 `current_value -= v_item`。
        更新 `current_j -= w_item`。
        继续下一个物品 `i-1`。
    *   **否则 (物品没有被选择)**：
        这意味着物品 `item` 没有被放入任何背包。`current_value` 应该等于 `dp[i-1][current_j][current_k]`。
        不需要更新 `current_j`, `current_k`, `current_value` (相对于 `i-1` 层的状态)。
        继续下一个物品 `i-1`。

3.  **返回结果**：
    *   返回最大价值、`bag1_items`、`bag2_items`。

注意：在回溯时，如果一个物品可以被放入任一背包都能达到 `dp[i][current_j][current_k]` 的值（例如，`value2 == value3 == dp[i][current_j][current_k]`），那么选择哪一个背包是任意的，这可能导致有多种合法的物品分配方案。通常，我们会按检查的顺序（例如，先检查背包2，再检查背包1）来确定。

````python
from typing import List, Tuple

def solve_two_knapsacks(V: List[int], W: List[int], c: int) -> Tuple[int, List[int], List[int]]:
    """
    解决双背包问题，返回最大总价值以及每个背包中的物品索引。
    V: 物品价值列表
    W: 物品重量列表
    c: 每个背包的容量
    """
    n = len(V)
    
    # dp[i][j][k] = 最大价值，考虑前i个物品，背包1容量j，背包2容量k
    # 初始化为0，因为价值是正的
    dp = [[[0 for _ in range(c + 1)] for _ in range(c + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        item_idx = i - 1 # 当前物品的0-based索引
        v_item = V[item_idx]
        w_item = W[item_idx]
        
        for j in range(c + 1): # 背包1的容量
            for k in range(c + 1): # 背包2的容量
                # Option 1: 不选择当前物品
                dp[i][j][k] = dp[i-1][j][k]
                
                # Option 2: 将当前物品放入背包1
                if j >= w_item:
                    dp[i][j][k] = max(dp[i][j][k], dp[i-1][j - w_item][k] + v_item)
                
                # Option 3: 将当前物品放入背包2
                if k >= w_item:
                    dp[i][j][k] = max(dp[i][j][k], dp[i-1][j][k - w_item] + v_item)

    # 找到最大的总价值
    max_value = 0
    final_j, final_k = 0, 0 # 记录达到最大价值时两个背包的容量
    for j in range(c + 1):
        for k in range(c + 1):
            if dp[n][j][k] > max_value:
                max_value = dp[n][j][k]
                final_j = j
                final_k = k
                
    # --- 回溯找到物品 ---
    bag1_items_indices = []
    bag2_items_indices = []
    
    current_j, current_k = final_j, final_k
    
    for i in range(n, 0, -1):
        item_idx = i - 1 # 当前物品的0-based索引
        v_item = V[item_idx]
        w_item = W[item_idx]
        
        # 检查物品是否来自不选它的情况
        # if dp[i][current_j][current_k] == dp[i-1][current_j][current_k]:
        #     continue # 物品i-1没有被选
        
        # 检查是否放入了背包2 (优先检查这个是为了匹配一种可能的选择顺序)
        # 注意：dp[i-1][current_j][current_k - w_item] 可能是0，如果 k - w_item < 0
        # 所以要确保 current_k >= w_item
        if current_k >= w_item and dp[i][current_j][current_k] == (dp[i-1][current_j][current_k - w_item] + v_item):
            bag2_items_indices.append(item_idx) # 存储0-based索引
            current_k -= w_item
        # 检查是否放入了背包1
        elif current_j >= w_item and dp[i][current_j][current_k] == (dp[i-1][current_j - w_item][current_k] + v_item):
            bag1_items_indices.append(item_idx) # 存储0-based索引
            current_j -= w_item
        # else: 物品 i-1 没有被选，dp[i][current_j][current_k] 应该等于 dp[i-1][current_j][current_k]
        # 这种情况由循环继续到 i-1 时的 dp[i-1][current_j][current_k] 自动处理

    # 题目示例中的物品下标似乎是1-based，并且按价值/重量排序（如果V和W相同）
    # 我们这里返回0-based索引，顺序是回溯时找到的逆序，可以reverse
    bag1_items_indices.reverse()
    bag2_items_indices.reverse()
            
    return max_value, bag1_items_indices, bag2_items_indices

# 示例测试
V_example = [1, 3, 2, 5, 8, 7]
W_example = [1, 3, 2, 5, 8, 7] # 假设物品重量和价值相同
c_example = 7

max_val, bag1, bag2 = solve_two_knapsacks(V_example, W_example, c_example)
print(f"输入 V={V_example}, W={W_example}, c={c_example}")
print(f"最大价值={max_val}")
# 为了匹配示例输出的1-based索引，我们可以在打印时转换
print(f"背包1装的物品 (0-based index): {bag1}")
print(f"背包2装的物品 (0-based index): {bag2}")

# 示例输出：最大价值=14，背包装的物品为：[6] [4，3]
# 对应0-based索引: [5] 和 [3, 2]
# 物品5: V=7, W=7. 放入背包1 (7) -> 背包1: [5], 背包2: []
# 物品3: V=5, W=5. 放入背包2 (5) -> 背包1: [], 背包2: [3]
# 物品2: V=2, W=2. 放入背包2 (2) -> 背包1: [], 背包2: [3, 2] (总重7，总价值7)
# 那么总价值是 7 (物品5) + 7 (物品3+2) = 14

# 调整打印以匹配1-based（如果需要）
print(f"背包1装的物品 (1-based index): {[x + 1 for x in bag1]}")
print(f"背包2装的物品 (1-based index): {[x + 1 for x in bag2]}")


V_example2 = [10, 40, 30, 50]
W_example2 = [5, 4, 6, 3]
c_example2 = 10
max_val2, bag1_2, bag2_2 = solve_two_knapsacks(V_example2, W_example2, c_example2)
print(f"\n输入 V={V_example2}, W={W_example2}, c={c_example2}")
print(f"最大价值={max_val2}")
print(f"背包1装的物品 (0-based index): {bag1_2}")
print(f"背包2装的物品 (0-based index): {bag2_2}")
# 预期：
# 物品0 (V10,W5), 物品1 (V40,W4), 物品2 (V30,W6), 物品3 (V50,W3)
# c=10
# Bag1: item 1 (V40,W4), item 3 (V50,W3) -> Total V=90, W=7
# Bag2: item 2 (V30,W6) -> Total V=30, W=6
# Total Value = 120

# your algorithm time complexity is: O(nc^2)
````

## 问题 10 

> 给定两个字符串 $x[1..n]$ 和 $y[1..m]$，我们想通过以下操作将 $x$ 变换为 $y$ :
>
> **插入**：在 $x$ 中插入一个字符(在任何位置)；**删除**：从 $x$ 中删除一个字符(在任何位置)； **替换**：用另一个字符替换 $x$ 中的一个字符。
>
> 例如: $x = abcd$, $y = bcfe$，
>
> - 将 $x$ 转换为 $y$ 的一种可能方法是：1. 删除 $x$ 开头的 $a$, $x$变成 $bcd$； 2. 将 $x$ 中的字符 $d$ 替换为字符 $f$。$x$ 变成 $bcf$； 3. 在 $x$ 的末尾插入字符 $e$。$x$ 变成 $bcfe$。
>
> - 另一种可能的方法：1. 删除 $x$ 开头的 $a$,  $x$ 变成 $bcd$； 2. 在 $x$ 中字符 $d$ 之前插入字符 $f$。$x$ 变成 $bcfd$。3. 将 $x$ 中的字符 $d$ 替换为字符 $e$。$x$ 变成 $bcfe$。

设计一个时间复杂度为 $O(mn)$ 的算法，返回将 $x$ 转换为 $y$ 所需的最少操作次数。分析算法设计思路，并基于python编程实现。


idea：

### 算法设计思路

这个问题旨在找到将字符串 `x` 转换为字符串 `y` 所需的最少操作次数，允许的操作有插入、删除和替换。这是一个经典的动态规划问题，通常使用 Wagner-Fischer 算法解决。

**1. 定义 DP 状态**

我们创建一个二维数组 `dp`，其中 `dp[i][j]` 表示将字符串 `x` 的前 `i` 个字符（即 `x[0...i-1]`，如果 `x` 是0索引的）转换成字符串 `y` 的前 `j` 个字符（即 `y[0...j-1]`）所需的最少操作次数。

DP 表的大小将是 `(n+1) x (m+1)`，其中 `n = len(x)` 且 `m = len(y)`。

**2. 初始化 DP 表**

*   `dp[0][0] = 0`:
    将一个空字符串转换成另一个空字符串需要0次操作。

*   `dp[i][0] = i` for `i` from 1 to `n`:
    将 `x` 的前 `i` 个字符转换成空字符串需要 `i` 次删除操作。
    例如，`x = "cat"`, `y = ""`. `dp[3][0]` (transform "cat" to "") = 3 (delete 't', delete 'a', delete 'c').

*   `dp[0][j] = j` for `j` from 1 to `m`:
    将空字符串转换成 `y` 的前 `j` 个字符需要 `j` 次插入操作。
    例如，`x = ""`, `y = "dog"`. `dp[0][3]` (transform "" to "dog") = 3 (insert 'd', insert 'o', insert 'g').

**3. 填充 DP 表（递推关系）**

我们从 `i = 1` 到 `n` 以及 `j = 1` 到 `m` 来填充 `dp` 表。
对于每个 `dp[i][j]`，我们考虑 `x` 的第 `i` 个字符（即 `x[i-1]`）和 `y` 的第 `j` 个字符（即 `y[j-1]`）。

*   **情况 1: 如果 `x[i-1] == y[j-1]`** (当前考虑的两个字符相同)
    在这种情况下，最后一个字符匹配，不需要为这两个字符执行任何操作。最少操作次数与将 `x` 的前 `i-1` 个字符转换为 `y` 的前 `j-1` 个字符的操作次数相同。
    `dp[i][j] = dp[i-1][j-1]`

*   **情况 2: 如果 `x[i-1] != y[j-1]`** (当前考虑的两个字符不同)
    我们需要执行一次操作（插入、删除或替换）来使它们匹配（或之一匹配前缀）。我们选择这三种操作中成本最低的一种：
    1.  **替换 (Replace)**: 将 `x[i-1]` 替换为 `y[j-1]`。
        成本是 `1` (替换操作) + `dp[i-1][j-1]` (将 `x[0...i-2]` 转换为 `y[0...j-2]` 的成本)。
    2.  **删除 (Delete)**: 从 `x` 中删除 `x[i-1]`。
        成本是 `1` (删除操作) + `dp[i-1][j]` (将 `x[0...i-2]` 转换为 `y[0...j-1]` 的成本)。
    3.  **插入 (Insert)**: 在 `x` (处理到 `x[i-1]` 之后) 中插入 `y[j-1]`。
        成本是 `1` (插入操作) + `dp[i][j-1]` (将 `x[0...i-1]` 转换为 `y[0...j-2]` 的成本)。

    所以，`dp[i][j] = 1 + min(dp[i-1][j-1], dp[i-1][j], dp[i][j-1])`

**4. 最终结果**

填充完整个 `dp` 表后，`dp[n][m]` 将包含将整个字符串 `x` (长度 `n`) 转换为整个字符串 `y` (长度 `m`) 所需的最少操作次数。

**时间复杂度分析**
*   我们需要填充一个大小为 `(n+1) x (m+1)` 的 DP 表。
*   计算每个单元格 `dp[i][j]` 需要常数时间（几次比较和一次 `min` 操作）。
*   因此，总的时间复杂度为 **O(n * m)**。

**空间复杂度分析**
*   我们使用了一个大小为 `(n+1) x (m+1)` 的 DP 表。
*   因此，空间复杂度为 **O(n * m)**。
    (这个空间复杂度可以优化到 O(min(n, m))，如果只需要最终结果而不需回溯路径，方法是只保留DP表中的当前行和前一行。)

````python
def min_edit_distance(x: str, y: str) -> int:
    """
    计算将字符串 x 转换为字符串 y 所需的最少操作次数 (编辑距离)。
    允许的操作：插入、删除、替换。
    """
    n = len(x)
    m = len(y)

    # dp[i][j] 是将 x 的前 i 个字符转换成 y 的前 j 个字符的最小编辑距离
    # dp 表的大小是 (n+1) x (m+1)
    dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

    # 初始化第一行和第一列
    for i in range(n + 1):
        dp[i][0] = i  # 将 x[0...i-1] 转换为空字符串需要 i 次删除
    for j in range(m + 1):
        dp[0][j] = j  # 将空字符串转换为 y[0...j-1] 需要 j 次插入

    # 填充 dp 表
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            char_x = x[i-1] # 当前 x 中的字符 (0-indexed)
            char_y = y[j-1] # 当前 y 中的字符 (0-indexed)

            if char_x == char_y:
                # 如果字符相同，则不需要操作，成本来自之前的子问题
                dp[i][j] = dp[i-1][j-1]
            else:
                # 如果字符不同，考虑三种操作：
                cost_replace = dp[i-1][j-1] # 替换 x[i-1] 为 y[j-1]
                cost_delete = dp[i-1][j]    # 从 x 中删除 x[i-1]
                cost_insert = dp[i][j-1]    # 在 x (处理到 x[i-1]后) 插入 y[j-1]
                
                dp[i][j] = 1 + min(cost_replace, cost_delete, cost_insert)
                
    return dp[n][m]

# 示例测试
x1 = "abcd"
y1 = "bcfe"
# 1. 删除 a -> bcd
# 2. 替换 d 为 f -> bcf
# 3. 插入 e -> bcfe
# Total 3 operations
output1 = min_edit_distance(x1, y1)
print(f"x = \"{x1}\", y = \"{y1}\", 最少操作次数: {output1}") # 期望: 3

x2 = "horse"
y2 = "ros"
# 1. replace h with r -> rorse
# 2. delete r -> rose
# 3. delete e -> ros
# Total 3
# OR
# 1. delete h -> orse
# 2. delete e -> ors
# 3. (no change for o, r, s)
# Actually:
# horse -> rorse (replace h by r) : 1
# rorse -> rose (delete r) : 1
# rose -> ros (delete e) : 1
# Total: 3
output2 = min_edit_distance(x2, y2)
print(f"x = \"{x2}\", y = \"{y2}\", 最少操作次数: {output2}") # 期望: 3

x3 = "intention"
y3 = "execution"

# Total 5
output3 = min_edit_distance(x3, y3)
print(f"x = \"{x3}\", y = \"{y3}\", 最少操作次数: {output3}") # 期望: 5

x4 = "sunday"
y4 = "saturday"

output4 = min_edit_distance(x4, y4)
print(f"x = \"{x4}\", y = \"{y4}\", 最少操作次数: {output4}") # 期望: 3


# your algorithm time complexity is: O(mn)
````
