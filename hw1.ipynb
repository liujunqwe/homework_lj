   当$n/4^k = 1$时迭代停止，即$4^k = n$，得到$k = \log_4 n$

2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$

   比较$f(n) = n^2$和$n^{\log_b a} = n^1 = n$

3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$

   取对数：$k \cdot \log(2/3) \leq -\log n$
   
   由于$\log(2/3) < 0$，所以$k \geq \frac{\log n}{-\log(2/3)} = \frac{\log n}{\log(3/2)}$

   所以适用主定理的第一种情况，得到$T(n) = \Theta(n^{\log_4 5}) = \Theta(n^{\log_4 5})$

5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9



idea:

````python
def max_profit_linear(profits):
    n = len(profits)
    
    # 初始化
    current_sum = 0
    max_sum = float('-inf')
    start = 0
    temp_start = 0
    end = 0
    
    # 使用Kadane算法找最大子数组和
    for i in range(n):
        # 如果当前和为负，重新开始计算
        if current_sum < 0:
            current_sum = profits[i]
            temp_start = i
        else:
            current_sum += profits[i]
        
        # 更新最大和及其对应的开始和结束位置
        if current_sum > max_sum:
            max_sum = current_sum
            start = temp_start
            end = i
    
    # 返回最大收益和对应的买入卖出日期（从1开始计数）
    return max_sum, start + 1, end + 1

# 测试
profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
profit, buy_day, sell_day = max_profit_linear(profits)
print(f"最大收益：{profit}，买入日：{buy_day}，卖出日：{sell_day}")
print("时间复杂度为O(n)")
````

````python
# add your code here
def majority_element_hashmap(nums):
    # 使用哈希表记录每个元素出现的次数
    count_map = {}
    n = len(nums)
    majority_threshold = n // 2
    
    for num in nums:
        # 如果元素不在哈希表中，将其加入并初始化计数为1
        # 否则将其计数加1
        count_map[num] = count_map.get(num, 0) + 1
        
        # 如果当前元素的计数超过了n/2，则它是多数元素
        if count_map[num] > majority_threshold:
            return num
    
    # 如果没有找到多数元素（虽然题目保证存在），返回None
    return None

# 测试
nums1 = [3, 2, 3]
nums2 = [2, 2, 1, 1, 1, 2, 2]
print(f"数组 {nums1} 的多数元素是: {majority_element_hashmap(nums1)}")
print(f"数组 {nums2} 的多数元素是: {majority_element_hashmap(nums2)}")
# your algorithm time complexity is :
print("时间复杂度为O(n)")
````

````python
# algorithm time complexity：O(n), space complexity:O(1)
def majority_element(nums):
    # 初始化候选元素和计数器
    candidate = nums[0]
    count = 1
    
    # 遍历数组，执行摩尔投票算法
    for i in range(1, len(nums)):
        # 如果计数为0，更新候选元素
        if count == 0:
            candidate = nums[i]
            count = 1
        # 如果当前元素等于候选元素，计数加1
        elif nums[i] == candidate:
            count += 1
        # 如果当前元素不等于候选元素，计数减1
        else:
            count -= 1
    
    return candidate

# 测试
nums1 = [3, 2, 3]
nums2 = [2, 2, 1, 1, 1, 2, 2]
print(f"数组 {nums1} 的多数元素是: {majority_element(nums1)}")
print(f"数组 {nums2} 的多数元素是: {majority_element(nums2)}")
# add your code here
print("时间复杂度为O(n),空间复杂度为O(1)")
````

## 问题 7：

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1


idea：### 算法思路

根据题目条件，A[1]>A[2]且A[n-1]<A[n]，我们可以利用二分查找的思想：

1. 如果A[mid]比它左边的元素A[mid-1]小，且比右边的元素A[mid+1]小，那么A[mid]就是一个局部最小值。
2. 如果A[mid]比它左边的元素A[mid-1]大，那么根据A[1]>A[2]，在左半部分A[1...mid-1]中一定存在一个局部最小值。
3. 如果A[mid]比它右边的元素A[mid+1]大，那么根据A[n-1]<A[n]，在右半部分A[mid+1...n]中一定存在一个局部最小值。

这种方法能找到一个局部最小值的原因是：
- A[1]>A[2]保证了如果没找到局部最小，往左找一定能找到
- A[n-1]<A[n]保证了如果没找到局部最小，往右找一定能找到
- 每次二分，我们都能在一侧确定存在局部最小

时间复杂度为O(logn)，因为每次递归都将问题规模减半。

````python
# add your code here
def find_local_minimum(A):
    # 索引从0开始，所以调整一下
    n = len(A)
    left = 0
    right = n - 1
    
    # 特殊情况处理
    if n == 1:
        return A[0]
    if A[0] < A[1]:
        return A[0]
    if A[n-1] < A[n-2]:
        return A[n-1]
    
    # 二分查找
    while left <= right:
        mid = (left + right) // 2
        
        # 确保mid-1和mid+1在有效范围内
        if mid > 0 and mid < n - 1:
            # 检查mid是否是局部最小
            if A[mid] < A[mid-1] and A[mid] < A[mid+1]:
                return A[mid]
            
            # 如果左边元素比mid小，在左半部分查找
            elif A[mid-1] < A[mid]:
                right = mid - 1
            
            # 否则在右半部分查找
            else:
                left = mid + 1
        # 边界情况处理
        elif mid == 0:
            if A[mid] < A[mid+1]:
                return A[mid]
            else:
                left = mid + 1
        else:  # mid == n-1
            if A[mid] < A[mid-1]:
                return A[mid]
            else:
                right = mid - 1
    
    # 如果没有找到局部最小（不应该发生）
    return None

# 测试
A = [9, 3, 7, 2, 1, 4, 5]
print(f"数组 {A} 中的一个局部最小值是: {find_local_minimum(A)}")

print("算法时间复杂度: O(logn)")
# your algorithm time complexity is :
````

idea:

### 算法思路

要找出所有局部最小值，需要遍历整个数组一次，检查每个位置是否满足局部最小的条件：

1. 对于第一个元素（索引0），如果它小于第二个元素，则它是局部最小值。
2. 对于最后一个元素（索引n-1），如果它小于倒数第二个元素，则它是局部最小值。
3. 对于中间元素（索引1到n-2），如果它比左右两个相邻元素都小，则它是局部最小值。

由于需要遍历整个数组，时间复杂度为O(n)。这是找出所有局部最小值的最优时间复杂度，因为最坏情况下数组中的每个元素都可能是局部最小值（例如交替的大小值序列）。

````python
# add your code here
def find_all_local_minimums(A):
    n = len(A)
    result = []
    
    # 检查第一个元素
    if n > 1 and A[0] < A[1]:
        result.append(A[0])
    
    # 检查中间元素
    for i in range(1, n-1):
        if A[i] < A[i-1] and A[i] < A[i+1]:
            result.append(A[i])
    
    # 检查最后一个元素
    if n > 1 and A[n-1] < A[n-2]:
        result.append(A[n-1])
    
    return result

# 测试
A = [9, 3, 7, 2, 1, 4, 5]
print(f"数组 {A} 中的所有局部最小值是: {find_all_local_minimums(A)}")

# 算法时间复杂度: O(n)
# your algorithm time complexity is :
print("算法时间复杂度: O(n)")
````

````python
# add your code here
def k_smallest_sort(arr, k):
    # 使用归并排序
    def merge_sort(arr):
        if len(arr) <= 1:
            return arr
        
        # 分解
        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        
        # 合并
        return merge(left, right)

    def merge(left, right):
        result = []
        i = j = 0
        
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        
        result.extend(left[i:])
        result.extend(right[j:])
        return result
    
    # 排序整个数组
    sorted_arr = merge_sort(arr)
    
    # 返回前k个元素
    return sorted_arr[:k]

# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(f"排序法：数组的前{k}个最小元素是: {k_smallest_sort(arr, k)}")

# 算法时间复杂度: O(n log n)
# 归并排序的时间复杂度为O(n log n)，然后取前k个元素的时间复杂度为O(k)
# 总时间复杂度为O(n log n + k) = O(n log n)
# your algorithm time complexity is :
print("时间复杂度为O(nlogn)")
````

````python
import heapq

def k_smallest_heap(arr, k):
    # 创建一个空堆
    heap = []
    
    # 逐个插入元素构建堆
    for num in arr:
        heapq.heappush(heap, num)
    
    # 提取k个最小元素
    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap))
    
    return result

# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print(f"堆方法：数组的前{k}个最小元素是: {k_smallest_heap(arr, k)}")
print("时间复杂度为O(nlogn)")
# 算法时间复杂度分析:
# 1. 逐个插入元素构建堆:
#    - 每次插入操作的时间复杂度为O(log i)，其中i是当前堆的大小
#    - 总时间复杂度为O(log 1 + log 2 + ... + log n) = O(log n!)
#    - 根据斯特林公式，log n! = O(n log n)
#    - 因此，逐个插入构建堆的时间复杂度为O(n log n)
# 2. 提取k个最小元素的时间复杂度为O(k log n)
# 总时间复杂度: O(n log n + k log n) = O(n log n)，当k ≤ n时
````

````python
import heapq

def k_smallest_huge_dataset(stream, k):
    '''
    处理超大数据集，找到前k小的元素
    参数:
        stream: 一个迭代器，代表数据流
        k: 需要找到的最小元素个数
    返回值:
        前k小的元素列表（已排序）
    '''
    # 读取前k个元素
    max_heap = []
    for _ in range(k):
        try:
            num = next(stream)
            # 使用负数转换成最大堆（Python的heapq默认是最小堆）
            heapq.heappush(max_heap, -num)
        except StopIteration:
            # 如果数据流中元素少于k个
            break
    
    # 处理剩余元素
    try:
        while True:
            num = next(stream)
            # 如果当前元素小于堆顶元素（最大元素）
            if num < -max_heap[0]:
                # 移除堆顶并添加当前元素
                heapq.heappop(max_heap)
                heapq.heappush(max_heap, -num)
    except StopIteration:
        # 数据流处理完毕
        pass
    
    # 将最大堆转换为排序后的列表
    result = [-heapq.heappop(max_heap) for _ in range(len(max_heap))]
    return sorted(result)

# 模拟超大数据集的测试
def simulate_large_dataset(arr):
    for num in arr:
        yield num

# 测试
arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
stream = simulate_large_dataset(arr)
print(f"超大数据集方法：数组的前{k}个最小元素是: {k_smallest_huge_dataset(stream, k)}")
print("时间复杂度为O(n log k)")

# 算法时间复杂度分析:
# 1. 构建初始最大堆: O(k)
# 2. 处理每个元素的时间复杂度: O(log k)
# 3. 总时间复杂度: O(k + (n-k) log k) = O(n log k)
# 空间复杂度: O(k)
````

## 问题 10：

给定一个包含m个**字符串**的数组A, 其中不同的字符串可能有不同的字符数, 但数组中所有字符串的字符总数为n。设计一个算法在 $O(n)$ 时间内对字符串进行排序, 分析算法设计方案, 计算其时间复杂度, 并基于python编程实现该算法。请注意, 假设字符串只包含"a","b",...,"z",

['a', 'ab', 'abc', 'b', 'ba', 'c']

note：

-  两个字符之间的比较可以考虑比较他们对应的ASCII码值；
- python中可以使用`ord("a")`返回字符 “a”对应的ASCII值

idea:

````python
# add your code here
def sort_strings_linear_time(A):
    """
    使用字典树对字符串数组进行排序，时间复杂度为O(n)
    参数：
        A: 字符串数组
    返回：
        排序后的字符串数组
    """
    # 字符集大小（a-z为26个字符）
    R = 26
    
    # 构建字典树（Trie）节点的类
    class TrieNode:
        def __init__(self):
            self.children = [None] * R
            self.is_end_of_word = False
            self.word = None
    
    # 创建新节点
    def get_node():
        return TrieNode()
    
    # 将字符转换为索引
    def char_to_index(ch):
        return ord(ch) - ord('a')
    
    # 在字典树中插入一个单词
    def insert(root, word):
        current = root
        for ch in word:
            index = char_to_index(ch)
            if not current.children[index]:
                current.children[index] = get_node()
            current = current.children[index]
        current.is_end_of_word = True
        current.word = word
    
    # 从字典树中按字典序收集所有单词
    def collect_words(node, result):
        if node.is_end_of_word:
            result.append(node.word)
        
        for i in range(R):
            if node.children[i]:
                collect_words(node.children[i], result)
    
    # 创建字典树的根节点
    root = get_node()
    
    # 将所有字符串插入字典树
    for word in A:
        insert(root, word)
    
    # 收集排序后的结果
    result = []
    collect_words(root, result)
    
    return result

# 另一种实现方式：基于MSD（Most Significant Digit）的基数排序
def msd_radix_sort(A):
    """
    使用MSD基数排序对字符串数组进行排序
    参数：
        A: 字符串数组
    返回：
        排序后的字符串数组
    """
    # 字符转索引
    def char_at(s, d):
        if d < len(s):
            return ord(s[d]) - ord('a') + 1  # +1确保空字符的索引为0
        return 0  # 字符串结束
    
    # 对A[lo..hi]的第d个字符进行排序
    def sort(a, lo, hi, d, aux):
        if hi <= lo:
            return
        
        # 计数数组（包括空字符）
        R = 27  # 26个字母加上字符串结束标记
        count = [0] * (R + 1)
        
        # 计算频率
        for i in range(lo, hi + 1):
            count[char_at(a[i], d) + 1] += 1
        
        # 转换为索引
        for r in range(R):
            count[r + 1] += count[r]
        
        # 分配
        for i in range(lo, hi + 1):
            aux[count[char_at(a[i], d)]] = a[i]
            count[char_at(a[i], d)] += 1
        
        # 回写
        for i in range(lo, hi + 1):
            a[i] = aux[i - lo]
        
        # 递归地对每个字符分组排序
        for r in range(R):
            sort(a, lo + count[r], lo + count[r+1] - 1, d + 1, aux)
    
    # 创建辅助数组
    aux = [None] * len(A)
    
    # 对整个数组进行排序
    sort(A, 0, len(A) - 1, 0, aux)
    
    return A

# 测试
test_cases = [
    ["a", "da", "bde", "ab", "bc", "abdc", "cdba"],
    ['ab', 'a', 'b', 'abc', 'ba', 'c'],
    ['aef', 'yzr', 'wr', 'ab', 'bhjc', 'lkabdc', 'pwcdba']
]

for i, test in enumerate(test_cases):
    test_copy = test.copy()
    print(f"测试用例 {i+1}:")
    print(f"输入: {test}")
    print(f"字典树排序后: {sort_strings_linear_time(test)}")
    print(f"MSD基数排序后: {msd_radix_sort(test_copy)}")
    print()


# your algorithm time complexity is :
````
